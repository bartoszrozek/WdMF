### Pierwszy sposób

```python
import time
```

```python
def generate_wiener(t):
    N = [0]
    for i in range(1,len(t)):
        N.append(N[-1] + math.sqrt(t[i] - t[i-1]) * np.random.normal(0,1) )
    return N
```

```python
start = time.time()
t = np.arange(0.1,15,0.01)
N = generate_wiener(t)
end = time.time()
print(end-start)
plt.plot(t,N)
plt.show()
```

### Drugi sposób

```python
from math import sqrt
def cholesky(A):
    n = len(A)

    L = [[0.0] * n for i in range(n)]

    for i in range(n):
        for k in range(i+1):
            tmp_sum = sum(L[i][j] * L[k][j] for j in range(k))
            
            if (i == k): 
                L[i][k] = sqrt(A[i][i] - tmp_sum)
            else:
                L[i][k] = (1.0 / L[k][k] * (A[i][k] - tmp_sum))
    return L
```

```python
start = time.time()
n = len(t)
Sigma = []
for i in range(n):
    Sigma.append((np.append(t[:i] ,np.repeat(t[i], n-i))))
    
A = cholesky(Sigma)
X = np.random.normal(0,1, len(t))
N = A @ X
end = time.time()
print(end-start)
```

```python
plt.plot(t,N)
plt.show()
```

Zbieżne z prawdopodobieństwem 1

```python
def generate_x(n):
    X = []
    for i in range(1,n+1):
        p = np.random.uniform(0,1)
        if p<1/(2*(i+1)**(3/2)):
            X.append(i+1)
        elif p<1/(1*(i+1)**(3/2)):
            X.append(-i-1)
        else:
            X.append(0)
    return X
```

```python
n=10000
for _ in range(20):
    X = generate_x(n)
    X = np.cumsum(X)
    X = [x/(i+1) for i,x in enumerate (X)]
    plt.plot(np.arange(1,n+1,1),X)
plt.show()
```

Zbieżne według prawdopodobieństwa

```python
def generate_y(n):
    X = []
    for i in range(1,n+1):
        p = np.random.uniform(0,1)
        if p<1/(2*(i+1)*math.log(i+1)):
            X.append(i+1)
        elif p<1/(1*(i+1)*math.log(i+1)):
            X.append(-i-1)
        else:
            X.append(0)
    return X
```

```python
n=10000
for _ in range(20):
    Y = generate_y(n)
    Y = np.cumsum(Y)
    Y = [y/(i+1) for i,y in enumerate (Y)]
    plt.plot(np.arange(1,n+1,1),Y)
plt.show()
```
